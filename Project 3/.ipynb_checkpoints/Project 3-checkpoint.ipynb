## 2. Data Preprocessing and Cleaning

```python
# drop datetime info
# -----------
chicago_inspections_2011_to_2013 = chicago_inspections_2011_to_2013.dropna().drop(['Inspection_Date',
                                                                                   'minDate',
                                                                                   'maxDate',
                                                                                   'Inspection_MonthYear'],
                                                                                  axis = 1)
```

Opted for the binary model for simplicity, but this is an arbitrary choice. This is an appropriate choice if city officials simply want to a simple way to prioritize inspections based on a Pass/Fail basis. 

City officials lose potentially useful information when the intermediate "pass w/conditions" category is eliminated. It is possible that there are features that are particular to that category but are potentially useful indicators of whether a business is closer to failing than passing or vice versa. That granularity is sacrificed in the interest of simplicity. 

```python
# process target 
# -----------
y = chicago_inspections_2011_to_2013['Results']

# decide if you want to binarize the outcome variable 
# -----------
# comment out the following lines of code if you don't want to binarize the target variable
y = y.replace({'Pass w/ Conditions': 'Pass'})
lb_style = LabelBinarizer()
y = lb_style.fit_transform(y)

# recode 0s and 1s so 1s are "Fail"
y = np.where(y == 1, 0 ,1)


# process features
# -----------

# create feature dataset
X = chicago_inspections_2011_to_2013.drop(columns = ['Results', 
                                                     'Facility_Type', 
                                                     'LICENSE_DESCRIPTION', 
                                                     'WARD_PRECINCT'])


# get dummies
X = pd.get_dummies(X)
```

I was interested in seeing why these other columns were dropped. ### Model 1: Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression 

# creating the model
logit_reg = LogisticRegression()

# Fitting the Model
logit_model = logit_reg.fit(X_train, y_train)

# predicting on validation data
y_pred = logit_model.predict(X_validate)
```

```python
# coefficent extraction for dataframe  
logit_data = pd.concat([pd.DataFrame(X.columns),
                        pd.DataFrame(np.transpose(logit_model.coef_))],
                       axis = 1)

logit_data.columns = ['Feature', 'Coefficient']
logit_data['abs_coef'] = abs(logit_data['Coefficient'])
```

```python
# plot 
sns.barplot(x="Coefficient", 
            y="Feature", 
            data=logit_data.nlargest(20, 'abs_coef')).set_title("Top Logit Coefficients")
plt.show()
```

#### Interpretation 


The logit regression suggests that the "serious count" feature has significant influence on whether a business passes or fails a health inspection. Each extra "serious" violation increases the log-odds of failing an inspection by 4 thus increasing the odds of failing by a factor of 54 (e^4). No other feature exerts such outsized influence on the "pass/fail" prediction. 

```python
# confusion matrix
cf_matrix = confusion_matrix(y_validate, 
                             y_pred, 
                             normalize = "true")

# dataframe conversion 
df_cm = pd.DataFrame(cf_matrix, range(2), range(2))

# figure specifications
df_cm = df_cm.rename(index=str, columns={0: "Pass", 1: "Fail"})
df_cm.index = ["Pass", "Fail"]
plt.figure(figsize = (10,7))
sns.set(font_scale=1.4)#for label size
sns.heatmap(df_cm, 
           annot=True,
           annot_kws={"size": 16},
           fmt='g')

# plot
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

The baseline model predicts true positives (i.e. businesses that pass inspections and are predicted to pass) and true negatives (i.e. businesses that fail inspections and are predicted to fail) with 94.1% and 86.4% respectively. 

The false negative rate is 5.8% whereby the model incorrectly classifies businesses that pass inspection as having failed. It also has a false positive rate of 13.6% and will incorrectly classify businesses that fail inspections as having passed at this rate.  

Were city officials to use this model without any additional tuning, they would need to decide whether it is safe to allow 13.5% of businesses that could potentially fail inspections to remain in operation. This means that when the model predicts 

## Hyperparameter Tuning

```python
# hyperparameter tuning

import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore')
from sklearn.metrics import accuracy_score

# parameters
param_grid = {'penalty': ['l1', 'l2', 'elasticnet'],
             'C': np.arange(.1, 1, .1),
             'fit_intercept': [True, False],
             'solver': ['liblinear', 'saga']}

# grid search
logit_grid = GridSearchCV(logit_model, 
                          param_grid, 
                          cv=3)
logit_grid.fit(X_train, 
               y_train)

# best model selection
best_index = np.argmax(logit_grid.cv_results_["mean_test_score"])
best_logit_pred = logit_grid.best_estimator_.predict(X_validate)

print(logit_grid.cv_results_["params"][best_index])
print('Validation Accuracy', accuracy_score(best_logit_pred, y_validate))
```

Hyperparameter tuning imposes both L1 (Lasso) and L2 (Ridge) penalties on the model. The first penalty (L1) reduce overfitting by driving non-essential (outlier) features to zero and the second penalty (L2), shrinks correlated features toward zero but does not eliminate them. 

The model's accuracy decreases to 93% post tuning of the validation cases with high accuracy. It is 93% accurate in making pass/fail determinations in the validation dataset. 

## Accuracy

This is a measure of model correctness i.e. how often is the model correct in its predictions. In this case, the model is 92.6% accurate when it predicts a business will either pass or fail an inspection.

```python
# Each object set to 0 to start
TP = 0
FP = 0
TN = 0
FN = 0

# loop over each observation to assign to TP, FP, TN, and FN
for i in range(len(y_pred)): 
    if y_validate[i]==y_pred[i]==1:
        TP += 1
    if y_pred[i]==1 and y_validate[i]==0:
        FP += 1
    if y_validate[i]==y_pred[i]==0:
        TN += 1
    if y_pred[i]==0 and y_pred[i]!=y_validate[i]:
        FN += 1
```

```python
# calculate accuracy based on the formula defined above
accuracy = (TP + TN)/(TP + TN + FP + FN)
print("Accuracy is", accuracy)
```

## Precision

This is a measure of how accurately the model predicts positive cases. So in this context, the model identifies businesses that pass inspections with 78.7% accuracy. In other words, of all the restaurants that the model predicted as passing inspections, 78.7% of them actually passed inspections. This also means that 21.3% of the businesses that were predicted to pass, actually failed the inspection. 

City officials must decide whether it is in the public's interest to tolerate a false positive rate of 21.3%. These are businesses that may continue to function while placing the public's health at risk. 



```python
precision = TP/(TP + FP)
print("Precision is", precision)
```

## Recall 

This is a measure of how many positive cases the model finds. So in this context, the model identifies 86.4% of businesses that pass inspections. This is a high recall rate. But it misclassifies 13.6% of businesses that pass inspections as failing. 

Misclassifying a business as failing an inspection likely results in adverse outcomes for that business. City officials must decide whether they are comfortable with dealing with the complaints, lawsuits, or other actions that those businesses may take and the resources the city is likely to expend to correct the misclassification. 

```python
recall = TP/(TP + FN)
print("Recall is", recall)
```

## F1 Score

This score attempts to manage the trade off between accurately predicting positive cases (precision) and locating the highest number of positive cases (Recall). An F1 score of 0.82 (close to 1) is high. It means that most businesses predicted to pass do pass, and businesses that pass are accurately identified by the model. 

City officials must decide whether this balance is acceptable for the public. They could raise the prediction threshold by making the criteria for passing much stricter. This would decrease the number of businesses that are incorrectly classified as passing which raises the precision rate. Increasing precision, however, risks missing businesses that should pass. 

City officials could alternatively lower the threshold for passing which would identify more businesses that pass inspections at the risks of increasing the rate of false positives. It is a trade off between allowing more businesses that are potentially unsafe to remain in operation against the disruption of compliant businesses. 


```python
f1 = 2 * (precision * recall)/(precision + recall)
print("F1 Score is", f1)
```

###  Model 2: Random Forest

Random forest builds multiple decision trees, each tree built (and then split) on a randomly selected subset of the training data and features from that data. Each tree makes an independent classification decision (i.e. whether a business passes or fails inspection in this context). Each tree makes a "pass/fail" prediction based on the subset of data and features it uses/observes. The model's final prediction is determined by majority. If most trees predict a "pass" for a restaurant then the overall model matches that prediction. 

The model's accuracy is determined by how well the model predicts a "pass or fail" on unseen, test data. 

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(
                       # specify parameters
                       n_estimators=100,              # specify the number of trees (100 is default)
                       criterion='gini',              # or you can use 'entropy' for information gain
                       max_depth=5,                   # how deep tree nodes can go
                       min_samples_split=2,           # samples needed to split node
                       min_samples_leaf=1,            # samples needed for a leaf
                       min_weight_fraction_leaf=0.0,  # weight of samples needed for a node
                       max_features=None,             # number of features to look for when splitting
                       max_leaf_nodes=None,           # max nodes
                       min_impurity_decrease=1e-07,   # early stopping
                       random_state = 10)             # random seed

# specify cross-validation
# ----------
scores = cross_val_score(rf_classifier, 
                         X, 
                         y.ravel(),
                         cv=5)

# average score across models
# ----------
scores.mean()
```

The mean score of 92.5 indicates that the random forest model accurately predicts pass/fail decisions about 92.5% of the time. This version of the model considers all of the features in the dataset

## Feature Importance

The "serious count" feature accounts for about 65% of the tree splitting decision. This finding appears to confirm the feature's importance/strength within the model. This finding also confirms the earlier finding in the logistic regression where the "serious count" feature increased the log-odds of failing an inspection by 4 which raises the odds of failing by a factor of 54 (e^4)) 

```python
# Random forest fit on data to identify feature importance 
rf_classifier.fit(X, y.ravel())

# feature importance dataframe
feat_importances = pd.concat([pd.DataFrame(X.columns),pd.DataFrame(np.transpose(rf_classifier.feature_importances_))], axis = 1)
feat_importances.columns = ["Feature", "Importance"]

# plot 
sns.barplot(x = "Importance", 
            y = "Feature", 
            data = feat_importances.nlargest(20, 'Importance')) 
plt.show()
```

### Model 3: ADABoost 

ADABoost is an ensemble approach that uses the "boosting" method where a sequence of decision trees attempt to correct the mistakes or misclassifications of preceding trees. Each new tree in the sequence learns from the mistakes of the preceding tree. In this case, the model places more weight on the samples that are more difficult to classify as either "pass" or "fail" with the goal of improving overall accuracy. 

```python
ada_classifier = AdaBoostClassifier(n_estimators=100)

# calculate accuracy using cross validation
# ----------
scores = cross_val_score(ada_classifier,  # classifier
                         X,               # specify features
                         y.ravel(),       # specify labels
                         cv = 5)          # specify 5-fold cross validation

# calculate mean score across models (i.e. decision trees?) 
# ----------
scores.mean()
```

The ADABoost score of 91.9 is a slight decrease from the random forest score of 92.5. This means that this model makes accurate pass/fail predictions approximately 91.9% of the time. Because the model does not use L1 (lasso) or L2 (ridge) regularization it is more likely to overfit the training data.

```python
# data processing
# -----------
chicago_inspections_2014 = pd.read_csv("data/Chicago Inspections 2014_updated.csv")

# drop unused features 
drop_cols = [
    'Results', 'Facility_Type', 'LICENSE_DESCRIPTION', 'WARD_PRECINCT',
    'AKA_Name', 'License', 'Address', 'City', 'State', 'Zip', 'Latitude', 'Longitude', 'Location',
    'ID', 'LICENSE_ID', 'LICENSE_TERM_START_DATE', 'LICENSE_TERM_EXPIRATION_DATE', 'LICENSE_STATUS',
    'ACCOUNT_NUMBER', 'LEGAL_NAME', 'DOING_BUSINESS_AS_NAME', 'ADDRESS', 'CITY', 'STATE', 'ZIP_CODE',
    'WARD', 'PRECINCT', 'LICENSE_CODE', 'BUSINESS_ACTIVITY_ID', 'BUSINESS_ACTIVITY', 'LICENSE_NUMBER',
    'LATITUDE', 'LONGITUDE', 'pass_flag', 'fail_flag', 'POLICE_DISTRICT',
    'Inspection_Date', 'minDate', 'maxDate', 'Inspection_MonthYear'
]
test_df = chicago_inspections_2014.drop(columns=drop_cols, errors='ignore').dropna()

# process features
X_test2014 = pd.get_dummies(test_df)
X_test2014 = X_test2014.reindex(columns=X.columns, fill_value=0)

# process target 
y_test2014 = chicago_inspections_2014.loc[X_test2014.index, 'Results']
y_test2014 = y_test2014.replace({'Pass w/ Conditions': 'Pass'})
lb_style = LabelBinarizer()
y_test2014 = lb_style.fit_transform(y_test2014)
y_test2014 = np.where(y_test2014 == 1, 0, 1)
```

```python
# predict and compare ADABoost
# -----------
y_pred = ada_classifier.predict(X_test2014)
y_pred_proba = ada_classifier.predict_proba(X_test2014)[:, 1]
```

```python
# evaluate
# -----------
print("Classification Report:\n", classification_report(y_test2014, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test2014, y_pred))
print("Accuracy Score:", accuracy_score(y_test2014, y_pred))
```

The ADABoost model correctly classified businesses as either failing or passing 90% of the time on unseen data. 95% of the businesses that the model classifies as passing inspections do pass them, and the model correctly identifies these businsses 92% of the time. It has a high F1 score of .94 suggesting a high balance of precision and recall. 

When the model predicts that a business will fail an inspection, it is only correct 72% of the time, and can only identify 80% of those businesses. This means that 28% of businesses classified as failing inspections actually pass their inspections. Thus while the false positive rate is low, the businesses that are incorrectly classified are likely to suffer or experience reputational or financial harm, or both.
